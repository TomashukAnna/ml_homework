**Задача 4**\
Сгенерировать массив из 20 случайных элементов, элементы которого лежат в интервале [0, 10].
+ Определить количество элементов, отличающихся от минимального на 5.
+ Найти среднее арифметическое всех четных элементов массива, стоящих на нечетных местах.

```python
arr = np.random.randint(0, 11, 20)

# задание 1
count = np.sum(np.abs(arr - np.min(arr)) == 5)
print(f'Число элементов, отличающихся от минимального на 5: {count}')

#задание 2
odd_places = arr[np.where(np.arange(len(arr)) % 2 != 0)]
print(f'Cреднее арифметическое всех четных элементов массива, стоящих на нечетных местах: {np.mean(odd_places)}')
```

**Задача 6**\
Сгенерировать матрицу B размером 5 на 5, элементы которой лежат в интервале [-5, 5]. Отрицательные элементы матрицы заменить на число отрицательных элементов в строке, в которой они находятся.

```python
arr = np.random.randint(-5, 5, (5, 5))
arr
```

```python
for i in range (arr.shape[0]):
  row = arr[i]
  n_row = np.sum(row < 0)
  row[row < 0] = n_row

print(arr)
```

```python
# извлечем список столбцов, по которым будем строить boxplot
column_names = df.columns.tolist()[2:]
types = df['School_Type'].unique()
```

```python
for column in column_names:    
    for school_type in types:
        plt.figure(figsize=(4, 3))        
        sns.boxplot(x='School_Type', 
                    y=column,                     
                    data=df[df['School_Type'] == school_type]
                    )        
        plt.title(f'Boxplot для {column} в {school_type}', fontsize=14)
        plt.xlabel('School Type', fontsize=12)        
        plt.ylabel(column, fontsize=12)
        plt.show()
```
